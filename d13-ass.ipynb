```python
import pandas as pd


df = pd.read_csv("C:/Users/amans/Downloads/Day_13_Pharma_data.csv")


print("Missing values per column:")
print(df.isnull().sum())


print("\nNumber of duplicate rows:")
print(df.duplicated().sum())


df = df.drop_duplicates()


df.fillna(df.mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot total sales per region
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=df, estimator='sum', errorbar=None, hue='Region', palette='viridis', legend=False)
plt.title('Total Sales per Region')
plt.ylabel('Total Sales')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
# Scatter plot: Marketing Spend vs Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df, color='blue')
plt.title('Marketing Spend vs Sales')
plt.xlabel('Marketing Spend')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Age_Group', y='Effectiveness', data=df, hue='Age_Group', palette='Set2', legend=False)
plt.show()
```

```python

# Calculate correlation matrix
corr_matrix = df[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
plt.title('Correlation Between Sales, Marketing Spend, and Effectiveness')
plt.tight_layout()
plt.show()
```
