```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("C:/Users/amans/Downloads/Day_9_banking_data.csv")


total_transaction_per_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()
total_transaction_per_account_type.plot(kind='bar', color='skyblue', title='Total Transaction Amount per Account Type')

plt.ylabel('Total Transaction Amount')
plt.xlabel('Account Type')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()

transaction_per_branch = df.groupby('Branch')['Transaction_Amount'].sum()
transaction_per_branch.plot(kind='pie', autopct='%1.1f%%', title='Percentage of Transactions per Branch', legend=False)

plt.ylabel('')  
plt.tight_layout() 
plt.show()
```
