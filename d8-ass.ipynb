```python
import pandas as pd


data = pd.read_csv('C:/Users/amans/Downloads/Day_7_sales_data.csv')


print("First 5 rows of the dataset:")
print(data.head())


print("\nBasic statistics of numerical columns:")
print(data.describe())


total_sales_by_region = data.groupby('Region')['Sales'].sum()
print("\nTotal sales for each region:")
print(total_sales_by_region)


most_sold_product = data.groupby('Product')['Quantity'].sum().idxmax()
print("\nMost sold product:", most_sold_product)


data['Profit Margin'] = (data['Profit'] / data['Sales']) * 100
average_profit_margin_by_product = data.groupby('Product')['Profit Margin'].mean()
print("\nAverage profit margin for each product:")
print(average_profit_margin_by_product)


sales_above_1000 = data[data['Sales'] > 1000]
print("\nRows where sales are greater than 1000:")
print(sales_above_1000)


sales_in_east = data[data['Region'] == 'East']
print("\nSales records for the 'East' region:")
print(sales_in_east)
```

```python
import pandas as pd


data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [1500, 800, 1200, 950],
    'Profit': [200, 100, 250, 150],
    'Quantity': [50, 40, 60, 30]
}

df = pd.DataFrame(data)


df['Profit_Per_Unit'] = df['Profit'] / df['Quantity']


df['High_Sales'] = df['Sales'].apply(lambda x: 'Yes' if x > 1000 else 'No')

print(df)
```
